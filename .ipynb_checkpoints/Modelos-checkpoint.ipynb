## Testando modelos - base Statlog (german credit data)

### Carregando a base e analisando estrutura:

```python
import pandas as pd
import numpy as np

import pydotplus 
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn import svm, tree
from sklearn.tree import export_graphviz
from scipy import stats
from sklearn.metrics import auc

from sklearn.neural_network import MLPClassifier

from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.metrics import roc_auc_score, average_precision_score, roc_curve

import seaborn as sns

from imblearn.over_sampling import SMOTE
```

```python
colunas = ['existing_checking_account','duration_month', 'credit_history', 'purpose', 'credit_amount', 'saving_account', 'present_employment_since', 'disposable_income_rate', 
           'personal_status_and_sex', 'debtors_guarantor','present_residence_since','property','age','installment_plans','housing','credits_this_bank',
           'job','people_provide_manutenance','telephone','foreign_worker', 'y']
```

```python
df = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/statlog/german/german.data', names = colunas, sep=' ', header = None)
df.head()
```

```python
catg_col = ['existing_checking_account', 'credit_history', 'purpose', 'saving_account', 'present_employment_since', 
           'personal_status_and_sex', 'debtors_guarantor','property','installment_plans','housing',
           'job','telephone','foreign_worker']
num_col = ['duration_month', 'credit_amount', 'disposable_income_rate','present_residence_since','age','credits_this_bank',
           'people_provide_manutenance', 'y']

for coluna in catg_col:
    df[coluna] = df[coluna].astype('category')
df.describe()
```

```python
# classe 1 = 1 e classe 2 = 0
df.y.replace([1,2], [1,0], inplace=True)
df.y.value_counts()
#base muito desbalanceada, necessidade de SMOTE
```

```python
from collections import defaultdict
from sklearn.dummy import DummyClassifier
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, label_binarize, StandardScaler, MinMaxScaler

# Labelencoding to transform categorical to numerical
# Enables better Visualization than one hot encoding
d = defaultdict(LabelEncoder)

# Encoding the variable
lecatdata = df[catg_col].apply(lambda x: d[x.name].fit_transform(x))

# print transformations
for x in range(len(catg_col)):
    print(catg_col[x],": ", df[catg_col[x]].unique())
    print(catg_col[x],": ", lecatdata[catg_col[x]].unique())
```

```python
dummyvars = pd.get_dummies(df[catg_col])
```

```python
df_clean = pd.concat([df[num_col], dummyvars], axis = 1)
```

```python
cols = list(df_clean.drop(['y'], axis=1).columns.values)
```

```python
X = df_clean.loc[:, df_clean.columns !='y']
y = df_clean.iloc[:, df_clean.columns == 'y']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3333, random_state=1)

y_train.y.value_counts()
```

#### SMOTE
Balanceando a base para que o número de ocorrências da classe 1 seja o mesmo da classe 2

```python
smt = SMOTE()
X_train_bal, y_train_bal = smt.fit_sample(X_train, y_train)
```

```python
np.bincount(y_train_bal)
```

##### Função para métricas

```python
def get_eval(clf, X_train, y_train,y_test,y_pred):
    # Cross Validation to test and anticipate overfitting problem
    scores1 = cross_val_score(clf, X_train, y_train, cv=10, scoring='accuracy')
    scores2 = cross_val_score(clf, X_train, y_train, cv=10, scoring='precision')
    scores3 = cross_val_score(clf, X_train, y_train, cv=10, scoring='roc_auc')
    # The mean score and standard deviation of the score estimate
    print("Cross Validation Accuracy: %0.2f (+/- %0.2f)" % (scores1.mean(), scores1.std()))
    print("Cross Validation Precision: %0.2f (+/- %0.2f)" % (scores2.mean(), scores2.std()))
    print("Cross Validation roc_auc: %0.2f (+/- %0.2f)" % (scores3.mean(), scores3.std()))
    # Create and print confusion matrix
    abclf_cm = confusion_matrix(y_test,y_pred)
    fig = plt.figure(figsize=(5,4))
    heatmap = sns.heatmap(abclf_cm, annot=True)
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    cm = pd.DataFrame(abclf_cm)
    print(cm)
    return
```

#### Função curva roc

```python
def curva_roc(y_test, y_pred):

    fpr, tpr, thresholds = roc_curve(y_test.values, y_pred, pos_label=1)

    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(5,5))
    plt.plot(fpr, tpr, color='darkorange', lw=1, label='ROC curve (area = %0.2f)' % roc_auc)
    #plt.plot(tpr, fpr, color='darkblue', lw=1, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Oharacteristic')
    plt.legend(loc="lower right")
    plt.show()
    return
```

### Árvore de Decisão

```python
def dectree_clf(X_train, y_train, X_test, y_test):
    print("Decision Tree CLF")
    
    dec_tree = tree.DecisionTreeClassifier(min_samples_split=10,min_samples_leaf=5).fit(X_train, y_train)
    
    y_pred = dec_tree.predict(X_test)
    
    get_eval(dec_tree, X_train, y_train,y_test,y_pred)
    
    curva_roc(y_test, y_pred)

    return
```

Primeiro com a base desbalanceada:

```python
dec_tree = dectree_clf(X_train, y_train, X_test, y_test)
```

Base balanceada:

```python
dec_tree_bal = dectree_clf(X_train_bal, y_train_bal, X_test, y_test)
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
def knnclf(X_train, y_train, X_test, y_test):  
    print("KNeighborsClassifier")
    knn = KNeighborsClassifier(n_neighbors=3).fit(X_train, y_train)

    # Predict target variables y for test data
    y_pred = knn.predict(X_test)

    # Get Cross Validation and Confusion matrix
    get_eval(knn, X_train, y_train,y_test,y_pred)
    curva_roc(y_test, y_pred)
    return
```

Base desbalanceada:

```python
j = y_train.values.ravel()
train_y = np.array(j).astype(int)
```

```python
knn = knnclf(X_train, train_y, X_test, y_test)
```

Base balanceada:

```python
knn_bal = knnclf(X_train_bal, y_train_bal, X_test, y_test)
```

### SVM (rbf e linear)

```python
 def linear_svm_clf(X_train, y_train, X_test, y_test, C): 
    print("LinearSVC")
    
    lin_svc = svm.LinearSVC(C=C, class_weight="balanced",dual=False).fit(X_train, train_y)

    y_pred = lin_svc.predict(X_test)
    
    get_eval(lin_svc, X_train, train_y,y_test,y_pred)
    curva_roc(y_test, y_pred)
    return
```

```python
def svm_rbf_clf(X_train, y_train, X_test, y_test, C): 
    print("SVC")
    #C = 0.7  # SVM regularization parameter
    #y = y_train.ravel()
    #train_y = np.array(y).astype(int)
    
    svc = svm.SVC(gamma ='scale' ,kernel='rbf',degree=1,probability=True,C=C,class_weight="balanced").fit(X_train, y_train)

    # Predict target variables y for test data
    y_pred = svc.predict(X_test)
    
    # Get Cross Validation and Confusion matrix
    get_eval(svc, X_train, y_train,y_test,y_pred)
    curva_roc(y_test, y_pred)
    return
```
